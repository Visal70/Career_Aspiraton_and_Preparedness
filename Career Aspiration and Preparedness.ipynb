```python
# Set seaborn style
sns.set(style='whitegrid')
```

```python
#Pie Charts
```

```python
# 1. Gender Distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
```

```python
# 2. Hometown Area Distribution
hometown_counts = df['Hometown_Area'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(hometown_counts, labels=hometown_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.axis('equal')
plt.show()
```

```python
# 3. Degree Type Distribution
degree_counts = df['Degree_Type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(degree_counts, labels=degree_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.axis('equal')
plt.show()
```

```python
# Objective 1: GPA Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['GPA'], kde=True, bins=10, color='skyblue')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Objective 2: Gender vs Career Goal
# Cross-tab: Gender vs Career Challenges
Career_goal_by_gender = pd.crosstab(df['Gender'], df['Career_Goal'])

# Convert to percentages
Career_goal_by_gender_percent = Career_goal_by_gender.div(Career_goal_by_gender.sum(axis=1), axis=0) * 100

# Plot
Career_goal_by_gender_percent.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.legend(title='Career Goal', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Objective 4: GPA vs Career preparedness level
plt.figure(figsize=(10, 6))

# Define custom order for x-axis
order = ['Very unprepared', 'Somewhat unprepared','Neither prepared nor unprepared' ,'Somewhat prepared', 'Very prepared']

sns.boxplot(
    data=df,
    x='Career_Preparedness',
    y='GPA',
    hue='Career_Preparedness',
    palette='coolwarm',
    legend=False,
    order=order 
)
plt.xlabel('Preparedness Level', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.xticks(rotation=45, fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# Objective 5: Degree Type vs Career Goal
# Cross-tab: Degree Type vs Career Goal
Career_goal_by_Degree_Type = pd.crosstab(df['Degree_Type'], df['Career_Goal'])

# Convert to percentages
Career_goal_by_Degree_Type_percent = Career_goal_by_Degree_Type.div(Career_goal_by_Degree_Type.sum(axis=1), axis=0) * 100

# Plot
Career_goal_by_Degree_Type_percent.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')
plt.xlabel('Degree Type')
plt.ylabel('Percentage')
plt.legend(title='Career Goal', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.xticks(rotation=45) 
plt.show()
```

```python
df1=df[['Career_Challenges','Hometown_Area']]
```

```python

df1.columns
```

```python
# Ensure df1 is a copy
df1 = df1.copy()

# Split only non-null values
df1['Career_Challenges'] = df1['Career_Challenges'].apply(
    lambda x: x.split(';') if pd.notnull(x) else x
)

# Explode the list to rows
df_exploded = df1.explode('Career_Challenges')

# Clean: remove spaces
df_exploded['Career_Challenges'] = df_exploded['Career_Challenges'].str.strip()

# Remove empty values
df_exploded = df_exploded[df_exploded['Career_Challenges'] != '']
print(df_exploded)
```

```python
df_exploded.head(20)
```

```python
# Objective 3: challenges vs hometown area
challenge_counts = df_exploded.groupby(['Hometown_Area', 'Career_Challenges']).size().reset_index(name='Count')

# Step 2: Plot grouped bar chart
plt.figure(figsize=(12, 6))

sns.barplot(data=challenge_counts, x='Hometown_Area', y='Count', hue='Career_Challenges')

plt.xlabel('Hometown Area')
plt.ylabel('Count')
plt.legend(title='Career Challenge', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside
plt.tight_layout()
plt.show()
```
